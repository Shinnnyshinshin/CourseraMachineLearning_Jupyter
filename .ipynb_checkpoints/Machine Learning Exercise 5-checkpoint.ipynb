#  Regularized Linear Regression and Bias vs Variance

## Regularized Linear Regression 

In the first half of the exerise, we will implement a regularized linear model that will predict the flow of water coming out of a dam using the change of water level in the resevoir.  

In the next half, you will go through some diagnostis of debugging algorithms and examine the effects of bias vs variance. 


### Visualizing the dataset

* X is the change in the water level
* Y is water flowing out of dam

The dataset is divided into 3 parts

* training set that your model will use to learn X and y
* cross validation set that will be used to determine the regularization parameter Xval yval 
* test set for evaluating performance. These are the ones that are unseen during the training set and are called Xtest and ytest


```python

a = si.loadmat('ex5_data/ex5data1.mat')
X = a['X']
y = a['y'].flatten()
Xval = a['Xval']
yval = a['yval'].flatten()
Xtest = a['Xtest']
ytest = a['ytest'].flatten()

# number of training examples
m = X.shape[0]
```

```python
# plot the training data
plt.plot(X, y, 'rx', mew=2)
plt.xlabel('Change in water level (x)')
plt.ylabel('Water flowing out of the dam (y)')
```

### Regularized linear regression cost function



```python
def linearRegCostFunction(X, y, theta, lam):
    
    m = len(y)
    
    lamarray = np.ones(len(theta)) * lam
    lamarray[0] = 0
    
    J = 0
    grad = np.zeros(theta.size)
    
    dt = np.dot(X, theta) - y
    
    J =  1 / 2 / m * (np.dot(dt, dt) + np.dot(lamarray, theta**2))
    
    grad = np.dot(X.T, (np.dot(X,theta) - y)) / m
    grad += lamarray * theta / m
    
    return J, grad
```

```python
# test cost function
theta = np.ones(2)
J, grad = linearRegCostFunction(np.hstack((np.ones((m, 1)), X)), y, theta, 1)
print (J, grad)
```

### Training the Linear Regression

```python
def trainLinearReg(X, y, lam):
    
    # initialize Theta
    initial_theta = np.zeros(X.shape[1])
    
    # create cost and grad functions
    cost = lambda x: linearRegCostFunction(X, y, x, lam)[0]
    grad = lambda x: linearRegCostFunction(X, y, x, lam)[1]
    
    # minimize using fmincg
    theta = fmin_cg(cost, initial_theta, fprime=grad, 
                    maxiter=200, disp=False)
    
    return theta

    :
```

```python

# run on training set
lam = 0
Xnew = np.hstack((np.ones((m, 1)), X))
theta = trainLinearReg(Xnew, y, lam)
```

```python
# plot results
plt.plot(X, y, 'rx', mew=2)
plt.plot(X, np.dot(Xnew,theta))
plt.xlabel('Change in water level (x)')
plt.ylabel('Water flowing out of the dam (y)')
```

Here we have set the regularization parameter to zero. . Because our current implementation of linear regression is trying to fit a 2 dimensional lambda, regularization will not be helpful (there will not be a problem of overfitting, because it isn't fitting all that well anyway.

## Bias-variance

An important concept in machine learning is the bias-variance trade-ff. Models wit high bias are not complex enough for the data and tend to underfit, while models with a high variance tend to overfit the training data. 


### Learning Curves

This is where learning curves come in. Learning  curve plots training and cross validation erros as a function of training-set size. We first need a training and cross validation set errror for different training set sizes. We will first generate subsets of the original training set X. 


```python
def learningCurve(X, y, Xval, yval, lam):
    
    m = X.shape[0]
    
    error_train = np.zeros(m)
    error_val = np.zeros(m)    
    
    # compute training error over subset of training set
    for ct, ii in enumerate(range(1,m+1)):
                
        # compute theta for traning subset
        theta = trainLinearReg(X[:ii,:], y[:ii], lam)
        
        # training/validation error
        error_train[ct] = linearRegCostFunction(X[:ii,:], y[:ii], theta, 0)[0]
        error_val[ct] = linearRegCostFunction(Xval, yval, theta, 0)[0]
        
    return error_train, error_val
```

```python
lam = 0
error_train, error_val = learningCurve(
    np.hstack((np.ones((m, 1)), X)), y,
    np.hstack((np.ones((Xval.shape[0], 1)), Xval)), 
    yval, lam)
```

```python
plt.plot(np.linspace(0, m, m), error_train, label='Train')
plt.plot(np.linspace(0, m, m), error_val, label='Cross Validation')
plt.xlabel('Number of training examples')
plt.ylabel('Error')
plt.legend()
```

We see that the number of traiining examples can increase but the number of training examples, but the error is still very high. This indicates a high bias in our mode, which means that the linear model is too simple can cannot fit the dataset well. 

## Polynomial Regression

The problem with our linear model was that it was too simple for the data and resulted in underfitting (high bias). In this part of the exercise, you will address this problem by adding more features.

\begin{equation*} 
     h_{\theta}(x) = \theta_{0} + \theta_{1}x_{1} + \theta_{2}x_{2} ..\theta_{p}x_{p} 
 \end{equation*}

Where xp is the water level to the power of P... we obtain a linear model where the features are various powers of the original value (water level)

```python
def polyFeatures(X, p):
    
    m = X.shape[0]
    Xpoly = np.zeros((m, p))
    Xpoly[:,0] = X[:,0]
    
    for ii in range(1,p):
        Xpoly[:,ii] = X[:,0]**(ii+1)
        
    return Xpoly

def featureNormalize(X):
    
    # compute mean and standard deviation
    mu = np.mean(X, axis=0)
    sigma = np.std(X, axis=0)
    
    X_norm = (X - mu) / sigma
    
    return X_norm, mu, sigma  

def plotFit(min_x, max_x, mu, sigma, theta, p):
    
    x = np.arange(min_x-15, max_x+25, 0.05)
    x = np.reshape(x, (len(x), 1))
    
    # Map X and normalize
    X_poly = polyFeatures(x, p)
    X_poly = (X_poly - mu) / sigma
    X_poly = np.hstack((np.ones((X_poly.shape[0], 1)), 
                         X_poly))
    
    # plot
    plt.plot(x, np.dot(X_poly,theta))
```

```python
p = 8

# Map X onto Polynomical Features and Normalize
X_poly = polyFeatures(X, p)
X_poly, mu, sigma = featureNormalize(X_poly)
X_poly = np.hstack((np.ones((m, 1)), X_poly))

# Map X_poly_test and normalize
X_poly_test = polyFeatures(Xtest, p)
X_poly_test = (X_poly_test - mu) / sigma
X_poly_test = np.hstack((np.ones((X_poly_test.shape[0], 1)), 
                         X_poly_test))

# Map X_poly_val and normalize
X_poly_val = polyFeatures(Xval, p)
X_poly_val = (X_poly_val - mu) / sigma
X_poly_val = np.hstack((np.ones((X_poly_val.shape[0], 1)), 
                         X_poly_val))
```

```python
# train model on training set
lam = 0
theta = trainLinearReg(X_poly, y, lam)

# learning curve
error_train, error_val = learningCurve(
    X_poly, y, X_poly_val, yval, lam)

# plot fit
plt.figure()
plt.plot(X, y, 'rx', mew=2)
plotFit(np.min(X), np.max(X), mu, sigma, theta, p)
plt.xlabel('Change in water level (x)')
plt.ylabel('Water flowing out of the dam (y)')
plt.title('Polynomial Regression Fit (lambda = %f)'%lam)

# plot learning curve
plt.figure()
plt.plot(np.linspace(0, m, m), error_train, label='Train')
plt.plot(np.linspace(0, m, m), error_val, label='Cross Validation')
plt.xlabel('Number of training examples')
plt.ylabel('Error')
plt.legend()
```

We see that the polynomial is able to fit the datapoints very very well. The training error is basically 0. However the polynomial is very complex and and goes off into infinity at the extreme values. These are all indicators of an overfitting model ... that will not generalize well. In our learning curve we see that there is a big gap between the training error and cross validation error, which is an indication of a high variance problem. 

### Adjusting Lambda value 

Let's see what happens to the values when start playing out with values of lambda. 

```python
# train model on training set
lams = [0, 1, 100]

for lam in lams:
    
    theta = trainLinearReg(X_poly, y, lam)

    # learning curve
    error_train, error_val = learningCurve(
        X_poly, y, X_poly_val, yval, lam)
    
    fig = plt.figure(figsize=(8,4))

    # plot fit
    ax = plt.subplot(121)
    ax.plot(X, y, 'rx', mew=2)
    plotFit(np.min(X), np.max(X), mu, sigma, theta, p)
    ax.set_xlabel('Change in water level (x)')
    ax.set_ylabel('Water flowing out of the dam (y)')

    # plot learning curve
    ax = plt.subplot(122)
    ax.plot(np.linspace(0, m, m), error_train, label='Train')
    ax.plot(np.linspace(0, m, m), error_val, label='Cross Validation')
    ax.set_xlabel('Number of training examples')
    ax.set_ylabel('Error')
    ax.legend()
    
    plt.suptitle('Polynomial Regression Fit (lambda = %f)'%lam)
```
